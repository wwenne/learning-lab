```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

import warnings
warnings.filterwarnings("ignore")

df_industries = pd.read_excel("Industry_Portfolios.xlsx")
df_market = pd.read_excel("Market_Portfolio.xlsx")
df_industries["Date"] = pd.to_datetime(df_industries["Date"],format = "%Y%m")
df_market["Date"] = pd.to_datetime(df_market["Date"],format = "%Y%m")
df_total = pd.merge(df_industries,df_market,how = "outer",on = "Date")
# get list of industry
list_industry = df_total.columns.tolist()[1:-1]
# get return deviation dataframe
df_deviation = df_total
for name in list_industry:
    df_deviation[name] = df_deviation[name] - df_deviation["Market"]
df_deviation_10 = df_deviation[['Date', 'NoDur', 'Durbl', 'Manuf',
                                'Enrgy', 'HiTec', 'Telcm', 'Shops',
                                'Hlth', 'Utils', 'Other']]
df_deviation_10 = df_deviation_10.set_index("Date")

expected_deviation_return = np.array(df_deviation_10.mean(axis = 0))

covariance_return = np.cov(df_deviation_10,rowvar = False)
reverse_covariance_matrix = np.linalg.inv(covariance_return)

unit_matrix = np.ones(len(expected_deviation_return))

alpha = np.dot(expected_deviation_return.T,np.dot(reverse_covariance_matrix,unit_matrix))
zeta = np.dot(expected_deviation_return.T,np.dot(reverse_covariance_matrix,expected_deviation_return))
delta = np.dot(unit_matrix.T,np.dot(reverse_covariance_matrix,unit_matrix))

Rmv = alpha / delta
Rp = np.arange(0,0.1,0.0005)

def tracking_err_frontier_portfolio(Rp):
    risk_square = \
        1/delta \
        + (delta/(zeta*delta-alpha**2))*(Rp-Rmv)**2
    risk = np.sqrt(risk_square)
    return risk

risk_p = tracking_err_frontier_portfolio(Rp)

plt.figure(figsize = (7,3))
plt.plot(risk_p,Rp,ls = "--",color = "b")
plt.xlabel("Tracking Error",fontsize = 10)
plt.ylabel("Expected Return Deviation",fontsize = 10)
plt.title("Minimum-Tracking-Error Frontier")
plt.show()
```

```python
# plot frontier with riskless asset
Rf = 0.0
def frontier_with_riskless_asset(Rp):
    risk_square = \
        ((Rp-Rf)**2) / (zeta - 2*alpha*Rf + delta*(Rf**2))
    risk = np.sqrt(risk_square)
    return risk

risk_p_with_riskless = frontier_with_riskless_asset(Rp)

plt.figure(figsize = (8,5))
# efficient frontier
plt.plot(risk_p,Rp,ls = "--",color = "b",label = "efficient frontier")
plt.xlabel("Tracking Error",fontsize = 10)
plt.ylabel("Expected Return Deviation",fontsize = 10)
plt.title("Minimum-Tracking-Error Frontier")

# tangent line
plt.plot(risk_p_with_riskless,Rp,linestyle = "-",color = "red",label = "tangent line")
plt.xlim([0,0.3])
plt.ylim([0,0.1])
plt.legend(loc = "upper left",bbox_to_anchor = (1,1))
plt.show()
```

```python
# Use Monte Carlo simulation to plot Frontier
df_2 = df_industries.set_index("Date")

num_simulations = 10**5
results = np.zeros((3,num_simulations))

for i in range(num_simulations):
    portfolio_weights = np.random.uniform(0,1,size = (10,1))
    portfolio_weights = portfolio_weights / np.sum(portfolio_weights)

    industry_return = np.array(df_2.mean())
    mean_return = np.dot(portfolio_weights.T,industry_return)

    industry_cov = np.cov(df_2,rowvar = False)
    covariance_return = np.dot(portfolio_weights.T,
                         np.dot(industry_cov,portfolio_weights))
    std_return = np.sqrt(covariance_return)
    
    results[0,i] = mean_return
    results[1,i] = std_return
    results[2,i] = mean_return / std_return

plt.figure(figsize = (12,7))
plt.scatter(results[1, :],results[0, :])
plt.title('Efficient Frontier')
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Return')
plt.show()
```
