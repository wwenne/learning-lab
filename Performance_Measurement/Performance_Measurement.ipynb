```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from sklearn import linear_model

import warnings
warnings.filterwarnings("ignore")

industry_df = pd.read_excel("Industry_Portfolios.xlsx")
industry_df["Date"] = industry_df["Date"].apply(lambda x: dt.datetime.strptime(str(x),"%Y%m"))
industry_df = industry_df.set_index("Date")
risk_factor_df = pd.read_excel("Risk_Factors.xlsx")
risk_factor_df["Date"] = risk_factor_df["Date"].apply(lambda x: dt.datetime.strptime(str(x),"%Y%m"))
risk_factor_df = risk_factor_df.set_index("Date")

industry_list = industry_df.columns.tolist()

# Sharpe Ratio

Sharpe_ratio_df = pd.DataFrame(index = ["Sharpe_ratio"])
for name in industry_list:
    expected_return = (industry_df[name] - risk_factor_df["Rf"]).mean()
    std_return = (industry_df[name] - risk_factor_df["Rf"]).std()
    Sharpe_ratio = expected_return / std_return
    Sharpe_ratio_df[name] = Sharpe_ratio
print(Sharpe_ratio_df)

# Sortino ratio (using risk-free rate as target)

Sortina_ratio_df = pd.DataFrame(index = ["Sortina_ratio"])
period = len(risk_factor_df)
for name in industry_list:
    summ = 0
    return_deviation = np.where((industry_df[name] - risk_factor_df["Rf"]) > 0,0,(industry_df[name] - risk_factor_df["Rf"]))
    summ = pd.DataFrame(return_deviation).apply(lambda x:x**2).sum()[0]
    expected_return = (industry_df[name] - risk_factor_df["Rf"]).mean()
    semi_variance = summ / period
    Sortina_ratio = expected_return / np.sqrt(semi_variance)
    Sortina_ratio_df[name] = Sortina_ratio
print(Sortina_ratio_df)

# Trenyor ratio

Trenyor_ratio_df = pd.DataFrame(index = ["Trenyor_ratio"])
for name in industry_list:
    y = industry_df[name] - risk_factor_df["Rf"]
    x = risk_factor_df[['Rm-Rf']]
    model = linear_model.LinearRegression()
    model.fit(x,y)
    Trenyor_ratio = model.coef_
    Trenyor_ratio_df[name] = Trenyor_ratio
print(Trenyor_ratio_df)

# Jensen's alpha

Jensen_alpha_df = pd.DataFrame(index = ["Jensen_alpha"])
for name in industry_list:
    expected_return = (industry_df[name] - risk_factor_df["Rf"]).mean()
    expected_market_return = risk_factor_df["Rm-Rf"].mean()
    model = linear_model.LinearRegression()
    model.fit(risk_factor_df[['Rm-Rf']],industry_df[name] - risk_factor_df["Rf"])
    beta_market = model.coef_
    Jensen_alpha = expected_return - beta_market * expected_market_return
    Jensen_alpha_df[name] = Jensen_alpha
print(Jensen_alpha_df)

# Three Factor Alpha

three_factor_alpha_df = pd.DataFrame(index = ["three_factor_alpha"])
for name in industry_list:
    independent = (industry_df[name] - risk_factor_df["Rf"])
    dependent = risk_factor_df[['Rm-Rf', 'SMB', 'HML']]
    model = linear_model.LinearRegression()
    model.fit(dependent,independent)
    three_factor_alpha = model.intercept_
    three_factor_alpha_df[name] = three_factor_alpha
print(three_factor_alpha_df)

# concat all dataframes
DF = pd.concat([Sharpe_ratio_df,Sortina_ratio_df,Trenyor_ratio_df,Jensen_alpha_df,three_factor_alpha_df]).T

fig,(ax1,ax2,ax3,ax4,ax5)= plt.subplots(5,figsize = (12,16))
ax1.bar(DF.index,DF["Sharpe_ratio"],color = "blue",label = "Sharpe Ratio",width = 0.5)
ax1.set_xlabel("Date")
ax1.set_ylabel("Sharpe Ratio")
ax1.set_title("Sharpe Ratio",fontsize = 14)
ax1.legend()
ax2.bar(DF.index,DF["Sortina_ratio"],color = "blue",label = "Sortina Ratio",width = 0.5)
ax2.set_xlabel("Date")
ax2.set_ylabel("Sortina Ratio")
ax2.set_title("Sortina Ratio",fontsize = 14)
ax2.legend()
ax3.bar(DF.index,DF["Trenyor_ratio"],color = "blue",label = "Trenyor Ratio",width = 0.5)
ax3.set_xlabel("Date")
ax3.set_ylabel("Trenyor Ratio")
ax3.set_title("Trenyor Ratio",fontsize = 14)
ax3.legend()
ax4.bar(DF.index,DF["Jensen_alpha"],color = "blue",label = "Jensen Alpha",width = 0.5)
ax4.set_xlabel("Date")
ax4.set_ylabel("Jensen Alpha(%)")
ax4.set_title("Jensen Alpha",fontsize = 14)
ax4.legend()
ax5.bar(DF.index,DF["three_factor_alpha"],color = "blue",label = "Three Factor Alpha",width = 0.5)
ax5.set_xlabel("Date")
ax5.set_ylabel("Three Factor Alpha(%)")
ax5.set_title("Three Factor Alpha",fontsize = 14)
ax5.legend()
plt.tight_layout()
plt.show()
```
